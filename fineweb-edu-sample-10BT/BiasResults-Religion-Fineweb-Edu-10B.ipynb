# Fineweb-Edu TF-IDF Religion Bias Analysis: Results

This displays the outputs from processing Fineweb-Edu sample 10BT for religion bias.

The columns show the top tf-idf values with respect to other words.

The columns with `+` in them show these same values, minus the mean tf-idf for all the other words each word co-occurs with. This provides a measure of how much the association is skewed with respect to the entire dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv('fineweb-edu-sample-10BT-religion-tfidf.csv', index_col=0)
print(df.columns)
```

```python
df[['word', 'muslim', 'muslim+', 'christian', 'christian+', 'jewish', 'jewish+', 'hindu', 'hindu+', 'buddhist', 'buddhist+', 'atheist', 'atheist+']][df['muslim+'] > 0].sort_values('muslim+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df[['word', 'christian', 'christian+', 'jewish', 'jewish+', 'hindu', 'hindu+', 'buddhist', 'buddhist+', 'atheist', 'atheist+', 'muslim', 'muslim+']][df['christian+'] > 0].sort_values('christian+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df[['word', 'jewish', 'jewish+', 'hindu', 'hindu+', 'buddhist', 'buddhist+', 'atheist', 'atheist+', 'muslim', 'muslim+', 'christian', 'christian+']][df['jewish+'] > 0].sort_values('jewish+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```
