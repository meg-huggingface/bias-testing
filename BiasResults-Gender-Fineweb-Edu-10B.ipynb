# Fineweb Edu TF-IDF Bias Analysis: Results
This displays the outputs from processing Fineweb edu data for bias analysis.

Here, we show the results for gender tf-idf for Fineweb Edu sample-10B.

The columns `man` and `woman` show the top tf-idf values with respect to other words.

The columns `man+` and `woman+` show these same values, minus the mean tf-idf for all the other words each word co-occurs with. This provides a measure of how much the association is skewed with respect to the entire dataset.


As can be seen, `man` is skewed towards the term `god`, and slightly skewed towards terms like `war`, `great`, and `king`.

`woman` is skewed towards terms like `pregnancy`, `cancer`, `mother`, `children`, and `family`.

```python
import pandas as pd
```

```python
df = pd.read_csv('fineweb-edu-sample-10BT-gender-tfidf.csv', index_col=0)
```

```python
df[['word', 'man', 'man+', 'woman', 'woman+', 'non-binary', 'non-binary+']].sort_values('man+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df[['word', 'woman', 'woman+', 'non-binary', 'non-binary+', 'man', 'man+']].sort_values('woman+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df[['word', 'non-binary', 'non-binary+', 'man', 'man+', 'woman', 'woman+']].sort_values('non-binary+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```
