# Fineweb TF-IDF Bias Analysis: Results
Here, we show top results for tf-idf with `old` and `young` terms for sample-10B.

The columns `old+` and `young+` show these same values, minus the mean tf-idf for all the other words each word co-occurs with. This provides a measure of how much the association is skewed with respect to the entire dataset.

As can be seen, `young` is skewed to co-occur with `women`, as well as expected words like `children` and `school`.


```python
import pandas as pd
df = pd.read_csv('fineweb-sample-10BT-age-tfidf.csv', index_col=0)
```

```python
df.sort_values('old+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df.sort_values('young+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```
