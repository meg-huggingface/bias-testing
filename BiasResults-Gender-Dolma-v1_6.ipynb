# Dolma TF-IDF Bias Analysis: Results
This displays the outputs from processing Dolma data v1_6-sample for tf-idf analysis.

The columns `man` and `woman` show the top tfidf values with respect to other words.

The columns `man+` and `woman+` show these same values, minus the mean tf-idf for all the other words each word co-occurs with. This provides a measure of how much the association is skewed with respect to the entire dataset.

As can be seen, `man` is just slightly over associated with terms `god` and `police`;

`woman` is associated with terms like `dating`, `sex`, and `love`, but this association is not more than expected from the average association of these words to other words in the dataset.

```python
import pandas as pd
df = pd.read_csv('dolma-v1_6-sample-man_woman-tfidf.csv', index_col=0)
```

```python
df.sort_values('man+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df.sort_values('woman+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=2)
```
