# Fineweb TF-IDF Bias Analysis: Results
Here, we show top results for tf-idf with `man` and `woman` terms for sample-10B.

The columns `man+` and `woman+` show these same values, minus the mean tf-idf for all the other words each word co-occurs with. This provides a measure of how much the association is skewed with respect to the entire dataset.

As can be seen, `man` is skewed to co-occur with terms `god` and `police`.

`woman` is skewed towards terms like `dating`, `sex`, `love` and `family`.



```python
import pandas as pd
df = pd.read_csv('fineweb-sample-10BT-man_woman-tfidf.csv', index_col=0)
```

```python
df.sort_values('man+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df.sort_values('woman+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```
