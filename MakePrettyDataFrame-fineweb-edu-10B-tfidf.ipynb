# Fineweb TF-IDF Bias Analysis: Results
This displays the outputs from processing Fineweb edu data for bias analysis.
Here, we show the results for tf-idf with `man` and `woman` terms for sample-10B.

As can be seen, `man` is skewed towards the term `god`, and slightly skewed towards terms like `war`, `great`, and `king`.

`woman` is skewed towards terms like `pregnancy`, `cancer`, `mother`, `children`, and `family`.

```python
import pandas as pd
```

```python
df = pd.read_csv('fineweb-edu-sample-10BT-man_woman-tfidf.csv', index_col=0)
```

```python
df.sort_values('man+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```

```python
df.sort_values('woman+', ascending=False).style.background_gradient(
    axis=None,
    vmin=0,
    vmax=0.2,
    cmap="YlGnBu"
).format(precision=3)
```
